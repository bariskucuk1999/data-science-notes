```python
#verileri görselleştirme
import numpy as np
import matplotlib.pyplot as plt
ageList=[10,20,30,30,30,40,50,60,70,75]
kilogramList=[20,60,80,85,86,87,70,90,95,90]
numpyList1=np.array(ageList)
numpyList2=np.array(kilogramList)
#numpy dizisini matplotlibe geçirmek istersek
plt.plot(numpyList1,numpyList2,"r") #(x axis, y axis, color)
plt.xlabel("Age")
plt.ylabel("Kilogram") #plt.show()
plt.title("Kilogram-Age Graph")
```

```python
numpyDizisi1=np.linspace(0,10,20)
numpyDizisi2=numpyDizisi1**3
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
plt.subplot(1,2,2) #(1 satır, 2 sütun, 2. grafik)
plt.plot(numpyDizisi2,numpyDizisi1,"b*-")
```

```python
myFigure=plt.figure()
figureAxes=myFigure.add_axes([0.2,0.2,0.9,0.9]) #(sol taraf eksen başl. noktası, alt taraf eksen başl. noktası, genişlik, uzunluk)
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni verileri")
figureAxes.set_ylabel("Y ekseni verileri")
figureAxes.set_title("Başlık")
```

```python
#iç içe grafikler
myFigure2=plt.figure()
figureAxes2=myFigure2.add_axes([0.1,0.1,0.7,0.7])
figureAxes3=myFigure2.add_axes([0.2,0.4,0.3,0.3]) #ilk 2 değer grafikte pozisyonlanmaya yarıyor
figureAxes2.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes2.set_xlabel("X ekseni verileri")
figureAxes2.set_ylabel("Y ekseni verileri")
figureAxes2.set_title("Ana Grafik Başlık")
figureAxes3.plot(numpyDizisi2,numpyDizisi1,"b")
figureAxes3.set_xlabel("X ekseni verileri")
figureAxes3.set_ylabel("Y ekseni verileri")
figureAxes3.set_title("Alt Grafik Başlık")
```

```python
(benimFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2)
#benimEksenler.plot(numpyDizisi1,numpyDizisi2,"b")
for eksen in benimEksenler:
  eksen.plot(numpyDizisi1,numpyDizisi2,color="#3A95A8",alpha=0.5,linewidth=4.0,linestyle="-.",marker="o",markersize=6,markerfacecolor="red")
  #alpha=şeffaflık, linewidth=kalınlık, linestyle=çizgi stili, marker=değer noktaları
  eksen.set_xlabel("X ekseni")
plt.tight_layout() #grafikler arası boşluk
```

```python
myFigure3=plt.figure(figsize=(4,4)) #dpi=80 (piksel olarak)
yeniEksen=myFigure3.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label="numpyDizisi**2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpyDizisi**3")
yeniEksen.legend(loc=2)
myFigure3.savefig("benimFigurum.png",dpi=80) #grafiği dosyalara kaydetme
```
