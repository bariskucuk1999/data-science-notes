**Seriler**

```python
import numpy as np
import pandas as pd
myDict={"Ahmet":40, "Kaan":56, "Yusuf":23}
pd.Series(myDict)
###veya###
myNames=["Ahmet","Kaan","Yusuf"]
myAges=[40,56,23]
pd.Series(myAges,myNames)#pd.Series(data,index)
yarismaSonucu=pd.Series([20,10,8],["Ahmet","Kaan","Yusuf"])
yarismaSonucu2=pd.Series([10,12,6],["Ahmet","Kaan","Yusuf"])
sonSonuc=yarismaSonucu+yarismaSonucu2 #serilerdeki değerleri toplama
```

**Data Frame**

```python
#data frame serilerin birleştirilmiş hali
data=np.random.randn(4,3) #4x3 random matrix
dataFrame=pd.DataFrame(data,index=["Ahmet","Kaan","Yusuf","Zehra"],columns=["Maas","Yas","CalismaSuresi"])
dataFrame[["Maas","Yas"]] #belirli column'lar getirildi
dataFrame.loc["Ahmet"] #belirli row getirildi
dataFrame.iloc[1] #index ile row getirildi
dataFrame["EmeklilikYasi"]=dataFrame["Yas"]+dataFrame["Yas"] #column'daki değerleri toplayıp yeni column oluşturma
dataFrame.drop("EmeklilikYasi",axis=1, inplace=True) #column silme(kalıcı silme)
dataFrame.drop("Kaan",axis=0) #row silme
dataFrame.loc["Yusuf","Yas"] #dataframe'den tek bir değeri alma
dataFrame[dataFrame["Yas"]<0] #dataframe'de sıfırdan küçük olan satırlar gösterilir
dataFrame.reset_index() #indexler numaralandırıldı
newIndexList=["TUR","TUR","IRA","BUL"]
dataFrame["Nationality"]=newIndexList #yeni column ekleme
dataFrame.set_index("Nationality") #yeni index tabloda başa geldi
```

# **Pandas Operasyonlar**

**Eksik Veri Getirme**

```python
#eksik veri getirme
myDict1={"Istanbul":[30,29,np.nan],"Ankara":[20,np.nan,25],"Izmir":[40,np.nan,np.nan]}
dataFrame=pd.DataFrame(myDict1)
dataFrame.dropna() #satırda NaN değer varsa o satırlar silinir
dataFrame.dropna(axis=1) #sütunda NaN değer varsa o sütunlar silinir
dataFrame.dropna(axis=1,thresh=2) #2 ve üstü NaN olan sütunlar silinir
dataFrame.fillna(20) #NaN değerler 20 ile dolduruldu
```

**DataFrame Gruplandırma**

```python
myDict2={"Departman":["Yazılım","Yazılım","Pazarlama","Pazarlama","İnsan Kaynakları","Hukuk"],
         "Calisanlar":["Ahmet","Ali","Burak","Sedat","Seyfi","Suna"],
         "Maas":[9000,5500,4500,7000,7500,8500]
        }
myDataFrame=pd.DataFrame(myDict2)
myGroupObject=myDataFrame.groupby("Departman")
myGroupObject.count() #hangi departmanda kaç kişi çalışıyor
```

```python
myGroupObject.mean() #departmandaki kişilerin ortalama maaşı
```

```python
myGroupObject.describe() #min, max gibi tüm özellikler hesaplanır 
```

**DataFrame Birleştirme**

```python
#verileri birleştirme
myDict3={"Isim":["Ahmet","Ali","Burak","Sedat"],
         "Spor":["Kosu","Yuzme","Kosu","Basketbol"],
         "Kalori":[800,650,450,300]
        }
dataFrame1=pd.DataFrame(myDict3,index=[0,1,2,3]) #data frame oluşturuldu 
myDict4={"Isim":["Bahri","Buket","Sezer","Yavuz"],
         "Spor":["Kosu","Yuzme","Kosu","Basketbol"],
         "Kalori":[750,650,950,400]
        }
dataFrame2=pd.DataFrame(myDict4,index=[4,5,6,7]) #data frame oluşturuldu 
myDict5={"Isim":["Naz","Necip","Vedat","Zafer"],
         "Spor":["Kosu","Yuzme","Badminton","Tenis"],
         "Kalori":[550,800,900,700]
        }
dataFrame3=pd.DataFrame(myDict5,index=[8,9,10,11]) #data frame oluşturuldu        
pd.concat([dataFrame1,dataFrame2,dataFrame3],axis=0) #column birleştirme olsa axis=1
```

```python
#tabloları birleştirme
myMergeDict1={"Isim":["Ahmet","Ali","Burak","Sedat"],
         "Spor":["Kosu","Yuzme","Kosu","Basketbol"],
        }
dataFrame4=pd.DataFrame(myMergeDict1) #data frame oluşturuldu 
myMergeDict2={"Isim":["Ahmet","Ali","Burak","Sedat"],
         "Kalori":[750,650,950,400]
        }
dataFrame5=pd.DataFrame(myMergeDict2) #data frame oluşturuldu
pd.merge(dataFrame4,dataFrame5,on="Isim") #ortak column üzerinden işlem
```

**Diğer Operasyonlar**

```python
myDataDict={"Departman":["Yazılım","Yazılım","Pazarlama","Pazarlama","İnsan Kaynakları","Hukuk"],
         "Calisanlar":["Ahmet","Ali","Burak","Sedat","Seyfi","Suna"],
         "Maas":[9000,5500,4500,7000,7500,8500]
        }
myDataFrame1=pd.DataFrame(myDataDict)
myDataFrame1["Departman"].unique() #tekrar eden değerler 1 defa gelir
myDataFrame1["Departman"].nunique() #unique değerlerin sayısı
myDataFrame1["Departman"].value_counts() #hangi departmandan kaç kişi olduğu
def brutMaas(maas):
  return maas*0.66
myDataFrame1["Maas"].apply(brutMaas) #fonksiyon ile maaş değerleri değiştirildi
myDataFrame1.isnull() #null değer olup olmadığına bakma
```

```python
#pivot table oluşturma
myDict6={"Spor":["Kosu","Kosu","Yuzme","Yuzme","Yuzme"],
         "Isim":["Ahmet","Ali","Burak","Sedat","Sedat"],
         "Yas":[35,32,25,40,29]
        }
kisiDataFrame=pd.DataFrame(myDict6)
kisiDataFrame
```

```python
kisiDataFrame.pivot_table(values="Yas",index=["Spor","Isim"],aggfunc=np.sum) #aynı isim için yaş değeri toplandı
```
