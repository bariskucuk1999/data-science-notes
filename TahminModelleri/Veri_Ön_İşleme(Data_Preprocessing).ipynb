```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
veriler=pd.read_csv("veriler.txt")
boyKilo=veriler[["boy","kilo"]]
boyKilo
```

**Eksik Verileri Tamamlamak**

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan, strategy="mean")
Yas=veriler.iloc[:,1:4].values
imputer=imputer.fit(Yas[:,1:4])
Yas[:,1:4]=imputer.transform(Yas[:,1:4])
Yas
```

**Kategorik Verilerin Encode Edilmesi (Sayısal Değere Dönüştürülmesi)**

```python
#nominal ve ordinal(kategorik veriler)-->numeric
ulke=veriler.iloc[:,0:1].values
from sklearn import preprocessing
le=preprocessing.LabelEncoder() #kategorik column dönüşümü için
ulke[:,0]=le.fit_transform(veriler.iloc[:,0])
ohe=preprocessing.OneHotEncoder() #column başlıklarını etiketlere taşıma ve her etiketin altında 0 veya 1 aitliğini belirleme
ulke=ohe.fit_transform(ulke).toarray() #tek aşamada fit&transform (önişlemedeki öğrenme ve dönüştürme süreci)
sonuc=pd.DataFrame(data=ulke, index=range(22),columns=["fr","tr","us"]) #numpy dizisi-->dataframe
sonuc2=pd.DataFrame(data=Yas, index=range(22),columns=["boy","kilo","yas"])
cinsiyet=veriler.iloc[:,-1].values
sonuc3=pd.DataFrame(data=cinsiyet,index=range(22),columns=["cinsiyet"])
s=pd.concat([sonuc,sonuc2],axis=1) #dataframe'lerin birleştirilmesi
s2=pd.concat([s,sonuc3],axis=1)
s2
```

**Verilerin Eğitim ve Test İçin Bölünmesi**

```python
#ulke, boy ve yas'tan cinsiyeti tahmin etme
from sklearn.model_selection import train_test_split #verinin satır bazlı bölünmesi
x_train, x_test, y_train, y_test=train_test_split(s,sonuc3,test_size=0.33,random_state=0)#verinin bağımlı ve bağımsız değişken olarak bölünmesi
x_train
y_train
#dikey eksende bağımlı ve bağımsız değişkenler olarak, yatay eksende train ve test(eğitim ve test) kümesi olarak ikiye ayırıyoruz)
```

**Sayısal Değerlerin Ölçeklenmesi**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(x_train)
X_test=sc.fit_transform(x_test)
#farklı anlamlardaki verilerin ortak anlama çekilmesi
#bir kişinin cinsiyetini tahmin etmede boy, yaş, kilonun etkisi nedir için ortalamasını alırsak belli bir değer olarak gelmesi(50-100 arası gibi)
```
