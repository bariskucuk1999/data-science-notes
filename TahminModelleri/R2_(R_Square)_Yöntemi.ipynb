```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
veriler=pd.read_csv("maaslar.txt")
x=veriler.iloc[:,1:2] #eÄitim seviyesi
y=veriler.iloc[:,2:] #maas
X=x.values #data frame bu Ã¶rnekte sklearn'de sÄ±kÄ±ntÄ± Ã§Ä±kardÄ±ÄÄ± iÃ§in numpy array dÃ¶nÃ¼ÅÃ¼mÃ¼
Y=y.values
regressor_dt = DecisionTreeRegressor(random_state=0)
regressor_dt.fit(X,Y)
Z = X + 0.5
K = X - 0.4
from sklearn.ensemble import RandomForestRegressor
rf_reg=RandomForestRegressor(n_estimators=10,random_state=0)
rf_reg.fit(X,Y.ravel())

#1 Ã§Ä±kmasÄ± Ã§ok iyi tahmin, 0 Ã§Ä±kmasÄ± kÃ¶tÃ¼ tahmin, negatif Ã§Ä±kmasÄ± Ã§ok kÃ¶tÃ¼ tahmin
#n--->kaÃ§ tane eleman olduÄu p--->kaÃ§ tane deÄiÅken aldÄ±ÄÄ±mÄ±zdÄ±r(1. denklemde 2 deÄiÅken, 2. denklemde 3 deÄiÅken)
from sklearn.metrics import r2_score
r2_score(Y,rf_reg.predict(X)) #r2_score(y_true,y_pred)
#Ã§Ä±kan sonuÃ§ random forest R2 deÄeri
```
