```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
veriler=pd.read_csv("maaslar.txt")
x=veriler.iloc[:,1:2] #eğitim seviyesi
y=veriler.iloc[:,2:] #maas
X=x.values #data frame bu örnekte sklearn'de sıkıntı çıkardığı için numpy array dönüşümü
Y=y.values
regressor_dt = DecisionTreeRegressor(random_state=0)
regressor_dt.fit(X,Y)
Z = X + 0.5
K = X - 0.4
#0.6, 1 ve 1.5 için aynı gruba indirildi(aynı sonuç olarak tahmin edildi)
#decision tree'de gelmiş olduğu aralık için aynı değeri döndürür
plt.scatter(X,Y,color="red")
plt.plot(x,regressor_dt.predict(X),color="blue")
plt.plot(x,regressor_dt.predict(Z),color="green")
plt.plot(x,regressor_dt.predict(K),color="yellow")
```

```python
regressor_dt.predict([[11]])
```

```python
regressor_dt.predict([[6.6]])
```

# **Random Forest(Rassal Ağaçlar)**

```python
#birden fazla decision tree'nin aynı veri kümesinde çizilmesi
#sonrasında problemin çözümünde hep birlikte kullanılması
#training ve prediction aşamaları, ensemble learning
from sklearn.ensemble import RandomForestRegressor
rf_reg=RandomForestRegressor(n_estimators=10,random_state=0) #n_estimators=>kaç tane decision tree çizileceği
rf_reg.fit(X,Y.ravel())
rf_reg.predict([[6.6]])
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,rf_reg.predict(X),color="blue")
plt.plot(X,rf_reg.predict(Z),color="green")
plt.plot(X,regressor_dt.predict(K),color="yellow")
```
