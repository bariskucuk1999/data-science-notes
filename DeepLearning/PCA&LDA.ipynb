**PCA(Principal Component Analysis)**
*   Boyut dönüştürme/indirgeme, gözetimsiz(sınıflar arası ayrım önemsiz)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
veriler=pd.read_csv("Wine.txt")
X=veriler.iloc[:,0:13].values
y=veriler.iloc[:,13].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=0)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
#boyut indirgeme(daha az sayıda column'a indirgeme)
from sklearn.decomposition import PCA #pca dönüşümü
pca=PCA(n_components=2)
X_train2=pca.fit_transform(X_train) #fit_transform()->hem eğitim hem uygulama
X_train2 #2 boyuta indirgendi
```

```python
X_test2=pca.transform(X_test) #eğitilmiş veriyi transform etme
X_test2
```

```python
#lojistik regresyon üzerinden sınıflandırma yapmak istersek;
#pca dönüşümünden önce gelen Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,y_train) #pca'dan çıkan bağımsız değişkenleri alıp bundan y_train'i öğrenme
#pca dönüşümünden sonra gelen Logistic Regression
classifier2=LogisticRegression(random_state=0)
classifier2.fit(X_train2,y_train)
#tahminler
y_pred=classifier.predict(X_test)
y_pred2=classifier2.predict(X_test2)
from sklearn.metrics import confusion_matrix
#actual/pca olmadan çıkan sonuç
print("Actual/Without PCA")
cm=confusion_matrix(y_test,y_pred)
print(cm)
#actual/pca sonrası çıkan sonuç
print("Actual/After PCA")
cm2=confusion_matrix(y_pred2,y_pred)
print(cm2)
#pca sonrası/pca öncesi
print("After PCA/Before PCA")
cm3=confusion_matrix(y_test,y_pred2)
print(cm3)
```

**LDA(Linear Discriminant Analysis)**
*   Boyut dönüştürme/indirgeme, gözetimli(sınıflar arası ayrım önemli)
*   Sınıflar arası farkı maximize etme




```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA(n_components=2)
X_train_lda = lda.fit_transform(X_train,y_train) #lda'nın çalışması için sınıfları öğrenmesi gerekli
#bu yüzden fit_transform() işlemine pca'dan farklı olarak y_train'i ekledik
X_train_lda
```

```python
X_test_lda=lda.transform(X_test)
X_test_lda
```

```python
#lda dönüşümünden sonra
classifier_lda=LogisticRegression(random_state=0)
classifier_lda.fit(X_train_lda,y_train)
#lda verisini tahmin etme
y_pred_lda=classifier_lda.predict(X_test_lda)
#lda sonrası/actual
print("After LDA/Actual")
cm4=confusion_matrix(y_pred,y_pred_lda)
print(cm4)
```
