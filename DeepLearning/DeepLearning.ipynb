```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
veriler=pd.read_csv("Churn_Modelling.txt")
veriler
```

```python
#veri ön işleme
X=veriler.iloc[:,3:13].values
Y=veriler.iloc[:,13].values
#encoding:kategorik->nümerik
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
X[:,1]=le.fit_transform(X[:,1]) #geography column
le2=preprocessing.LabelEncoder()
X[:,2]=le2.fit_transform(X[:,2]) #gender column
#one hot encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ohe=ColumnTransformer([("ohe",OneHotEncoder(dtype=float),[1])],remainder="passthrough")
X=ohe.fit_transform(X)
X=X[:,1:]
X
```

```python
X[0:5,:] #ilk 5 satırı yazdırma
```

```python
X[:,8:10] #8 ve 9. sütunları yazdırma
```

```python
#verilerin eğitim ve test için bölünmesi
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.33,random_state=0)
#verilerin ölçeklenmesi
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(x_train) #normalize edildi
X_test=sc.fit_transform(x_test)
X_train
```

**Müşteri Kayıp Analizi**

```python
#yapay sinir ağları
import keras
from keras.models import Sequential
from keras.layers import Dense
classifier=Sequential() #yapay sinir ağı oluşturuldu
classifier.add(Dense(6,kernel_initializer="uniform",activation="relu",input_dim=11)) #giriş katmanı(11 bağımsız değişken için)
#input_dim->yapay sinir ağı giriş katmanındaki veri sayısı
#6 nörondan oluşan bir gizli katman
classifier.add(Dense(6,kernel_initializer="uniform",activation="relu")) #ikinci gizli katmanı ekleme
```

```python
classifier.add(Dense(1,kernel_initializer="uniform",activation="sigmoid")) #çıkış katmanı
classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
classifier.fit(X_train,y_train,epochs=50) #x'den y'yi öğrenme(bağımsız değişkenlerden bağımlı değişkeni öğrenme)
y_pred=classifier.predict(X_test)
y_pred
```

```python
y_pred=(y_pred>0.5) #true olanları döndürme
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
#gerçek veriden elde edilen actual datalar ile elde ettiğimiz y_pred değerlerini karşılaştırma
cm
```
